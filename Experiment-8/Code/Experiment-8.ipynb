```python
# Import OpenCV for image processing
# Import NumPy for numerical operations
import cv2
import numpy as np
```

```python
# Load image from given path
img = cv2.imread('/content/img.jpg')
```

```python
# Convert image to grayscale (used for intensity-based operations)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Convert image from BGR to HSV color space
# HSV is better for color segmentation
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
# Split HSV into Hue, Saturation, and Value channels
h,s,v = cv2.split(hsv)
```

```python
# Print Hue values (represents color type)
print(h)
```

```python
# Print Saturation values (color intensity)
print(s)
```

```python
# Print Value channel (brightness)
print(v)
```

```python
# Define Hue range for GREEN color
lower_h = 35
upper_h = 85
```

```python
# Create mask for green pixels using Hue channel
mask = cv2.inRange(h, lower_h, upper_h)
```

```python
# Apply mask to original image to extract green regions
result = cv2.bitwise_and(img, img, mask=mask)
```

```python
# Save the result image for green detection
cv2.imwrite('result.jpg',result)
```

```python
# Define Hue range for RED (upper range due to HSV wrap-around)
lower_h2 = 160
upper_h2 = 180
```

```python
# Mask for red color (high hue range)
mask2 = cv2.inRange(h, lower_h2, upper_h2)
```

```python
# Extract red regions
result2 = cv2.bitwise_and(img, img, mask=mask2)
```

```python
# Save red detection result
cv2.imwrite('result2.jpg',result2)
```
