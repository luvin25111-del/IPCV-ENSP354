```python
import cv2
import numpy as np
```

```python
img = cv2.imread('/content/img.jpg')
```

```python
rows,cols = img.shape[:2]
```

```python
M_trans = np.float32([[1, 0, 100], [0, 1, 100]])
```

```python
translate = cv2.warpAffine(img, M_trans, (cols, rows))
```

```python
cv2.imwrite('translation.jpg',translate)
```

```python
scaled = cv2.resize(img, None, fx=0.2, fy=0.2)
```

```python
cv2.imwrite('scaled.jpg',scaled)
```

```python
M_rot = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
```

```python
rotated = cv2.warpAffine(img,M_rot,(cols,rows))
```

```python
cv2.imwrite('rotated.jpg',rotated)
```

```python
M_rot2 = cv2.getRotationMatrix2D((cols/3,rows/3),90,1)
```

```python
rotated2 = cv2.warpAffine(img,M_rot2,(cols,rows))
```

```python
cv2.imwrite('rotated2.jpg',rotated2)
```
